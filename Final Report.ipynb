<h1> Final Project - Finding a gym business the best location in Toronto

<h2>Introduction/Business Problem

For this project a business is looking for a suitable location to put their new gym business. They are going to build the new gym in the scarbrough area but need to find which part of scarbrough they wish to put it in. 

To do this the foursquare API will be used to find the most common venues in each area. These will then be used to rule out areas which already have a large number of gyms.
Venues which would reduce the need for a gym will also be considered and any area which has one of the following as their most common venues will be excluded from the final list
<li> Park
<li> Yoga Studio
<li> Field

After the most common venues have been considered the remaining list will be compared to see which has the largest population and therefore the largest target market for the new gym

To help aid the business, clustering will also be used to section areas off and to give the business a better idea of the area

<h3> Exploratory Data Analysis

As each area needs to be explored so that we can group areas of scarbrough to find the best area to locate the gym some exploratory analysis was needed to help give the business the best idea of the area.

A range of libaries were used for this such as
<li> Pandas

<li> matplotlib
    
Folium was used to create a map of toronto and create markers for each area of toronto to display the potential areas that the business could be placed in.
The Foursquare API was then used to find the venues at each area of toronto, and more specifically to find each area of toronto's most popular types of venues.

<h3>Analyzing Each Neighbourhood

As explained earlier the foursquare API is used to retrieve information on the venues that are in each neibourhood. A function is used to collect the top 10 venue types in each of the neighbourhoods from the API and these are added each to a new column to be able to further analyse the data.

<h3> Clustering the Neighbourhoods

K-means Machine learning was used here to cluster the neighbourhoods together and then a folium map as used to display this using a rainbow code folium map.

<h3> Creating a list of suitable, potential new gym locations

Firstly a new data set that only contained area in scarborough was created.
Then for a location to be suitable the first condition was that it needed to be in an area which didn't already have a large number of gyms. So any area which had a common venue of gym's was removed from the dataframe.
The dataframe then had rows removed which had a lot of venues that would reduce the need for a gym which were:
<li> Park
<li> Yoga Studio
<li> Field
    
This then left us with a list of potentially suitable neighbourhoods to put the new gym in.

<h3> Comparing Populations of suitable areas

A csv file was created from the a wikipedia page containing the population of each set of neighbourhoods. This was then merged with a condenced version of the earlier dataframe. That dataframe was then sorted to show the highest population areas that are suitable locations to help give the best recommendations for the business problem

<h2> Results

The results of the analysis showed that the best area to locate the a new gym in scarborough is in the Malvern and Rouge area as this had the largest population of any area that had little competition and therefore the largest amount of potential customers.
